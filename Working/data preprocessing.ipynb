```python
import pandas as pd

df=pd.read_csv('C:/Users/admin/OneDrive/Desktop/Symptoms_based_disease_prediction/Datasets/symbipredict_2022.csv')
print(df.shape)
df.head(2)
```

```python
# Number of unique diseases
print("Number of unique disease:",df['prognosis'].nunique())

# Check total number of duplicated rows
print("Total no. of duplicated rows:",df.duplicated().sum())
print("Shape of dataframe before duplicating:",df.shape)

# Remove duplicated rows
df = df.drop_duplicates()
print("Shape of dataframe after duplicating:",df.shape)
```

```python
# Get all columns except the last one (assuming 'prognosis' is the last column)
columns_to_check = df.columns[:-1]

# Create the new column
df['symptoms'] = df.apply(
    lambda row: ', '.join([col for col in columns_to_check if row[col] == 1]), axis=1
)
df.head()
```

```python
preprocessed_df=df[['symptoms','prognosis']]

preprocessed_df.to_csv('C:/Users/admin/OneDrive/Desktop/Symptoms_based_disease_prediction/Datasets/preprocessed_data.csv',index=False)   # Save the DataFrame to a CSV file
print("CSV file 'updated_records.csv' saved successfully.")
preprocessed_df.head()
```
